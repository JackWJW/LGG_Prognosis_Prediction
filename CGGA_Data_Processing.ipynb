```python
import pandas as pd
import numpy as np
```

```python
CGGA_FPKM = pd.read_csv("./CGGA_Data/CGGA.mRNAseq_693.RSEM-genes.20200506.txt", sep='\t')
CGGA_FPKM = CGGA_FPKM.set_index("Gene_Name")
CGGA_TPM = CGGA_FPKM.div(CGGA_FPKM.sum(axis=0), axis=1) * 1e6
CGGA_TPM = np.log2(CGGA_TPM+0.001)
CGGA_TPM = CGGA_TPM.T.copy()
CGGA_TPM.head()
```

```python
GENE_LIST = ['A4GALT', 'ABO', 'B3GALNT1', 'B3GALT1', 'B3GALT4', 'B3GALT5', 
    'B3GNT2', 'B3GNT5', 'B4GALNT1', 'B4GALT5', 'B4GALT6', 
    'FUT1', 'FUT2', 'FUT3', 'FUT5', 'FUT6', 'FUT9', 
    'GAL3ST1', 'GCNT2', 'ST3GAL1', 'ST3GAL2', 'ST3GAL3', 'ST3GAL4', 
    'ST3GAL5', 'ST3GAL6', 'ST6GALNAC2', 'ST6GALNAC3', 'ST6GALNAC4', 'ST6GALNAC5', 
    'ST6GALNAC6', 'ST8SIA1', 'ST8SIA5', 'UGCG', 'UGT8']
CGGA_TPM = CGGA_TPM[GENE_LIST]
CGGA_TPM.head()
```

```python
clinical_data = pd.read_csv("./CGGA_Data/CGGA.mRNAseq_693_clinical.20200506.txt", sep='\t')
clinical_data = clinical_data.set_index("CGGA_ID")
clinical_data.head()
```

```python
CGGA_TPM_Clinical = pd.merge(clinical_data, CGGA_TPM, left_index=True,right_index=True)
```

```python
LGG_CGGA = CGGA_TPM_Clinical[CGGA_TPM_Clinical["Grade"].isin(["WHO II", "WHO III"])]
```

```python
LGG_CGGA = LGG_CGGA.drop(columns=['PRS_type', 'Histology', 'Grade', 'Gender', 'Age',
                                    'Radio_status (treated=1;un-treated=0)',
       'Chemo_status (TMZ treated=1;un-treated=0)', 'IDH_mutation_status',
       '1p19q_codeletion_status', 'MGMTp_methylation_status'])
```

```python
LGG_CGGA.rename(columns={"OS":"OS.time","Censor (alive=0; dead=1)":"OS"},inplace=True)
LGG_CGGA.head()
```

```python
LGG_CGGA.to_csv("./CGGA_Data/CGGA_LGG_TPM.csv")
```
